### Is there a relationship between gender and specialization? (i.e. Does the preference of Specialisation depend on the Gender?


```python
from scipy.stats import chi2_contingency

# List of specialization columns
specialization_columns = ['computerscience', 'mechanicalengg', 'electricalengg', 'telecomengg', 'civilengg']

# ate a new column indicating specialization based on highest score in any engineering field
df['specialization'] = df[specialization_columns].idxmax(axis=1)

# Filter out cases where no specialization is indicated
df = df[df['specialization'] != -1]

# Create a contingency table for gender and specialization
contingency_table = pd.crosstab(df['gender'], df['specialization'])

print("Contingency Table for Gender and Specialization:")
print(contingency_table)

# Perform the Chi-Square Test
chi2_stat, p_val, dof, ex = chi2_contingency(contingency_table)

print(f"Chi-Square Statistic: {chi2_stat}, P-value: {p_val}")
```

```python
# Visualize the distribution with a countplot
plt.figure(figsize=(10, 6))
sns.countplot(x='specialization', hue='gender', data=df)
plt.title('Specialization Preferences by Gender')
plt.xlabel('Specialization')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
